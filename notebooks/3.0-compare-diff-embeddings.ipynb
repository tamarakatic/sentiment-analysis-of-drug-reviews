```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
bert_vs_elmo = pd.read_csv("../reports/bert_vs_elmo_report.csv")
```

```python
plt.figure(figsize=(6, 8))
sns.barplot([("{} ({:.2f}%)").format(row['method'], row['f1_micro']) for i, row in bert_vs_elmo.iterrows()], bert_vs_elmo['f1_micro'])


plt.title("BERT vs ELMo")
plt.xlabel("Model")
plt.xticks(rotation='vertical')
plt.ylabel("F1-micro")
plt.show()
```

```python
glove_fasttext = pd.read_csv("../reports/flair+glove_vs_flair+fasttext_report.csv")
```

```python
plt.figure(figsize=(6, 8))

sns.barplot([("{} ({:.2f}%)").format(row['method'], row['f1_micro']) for i, row in glove_fasttext.iterrows()], glove_fasttext['f1_micro'])

plt.title("Flair+Glove vs Flair+FastText")
plt.xlabel("Model")
plt.xticks(rotation='vertical')
plt.ylabel("F1-micro")
plt.show()
```

```python
embeddings = pd.read_csv("../reports/bert_elmo_glove_fasttext_report.csv")
```

```python
plt.figure(figsize=(6, 8))

sns.barplot([("{} ({:.2f}%)").format(row['method'], row['f1_micro']) for i, row in embeddings.iterrows()], embeddings['f1_micro'])

plt.title("BERT, Elmo, Flair+Glove and Flair+FastText")
plt.xlabel("Model")
plt.xticks(rotation='vertical')
plt.ylabel("F1-micro")
plt.show()
```

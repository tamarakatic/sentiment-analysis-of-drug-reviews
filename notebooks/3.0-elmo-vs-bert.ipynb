```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
bert_vs_elmo = pd.read_csv("../reports/bert_vs_elmo_report.csv")
```

```python
plt.figure(figsize=(6, 8))
sns.barplot(bert_vs_elmo['method'], bert_vs_elmo['f1_micro'])

plt.title("BERT vs ELMo")
plt.xlabel("Model")
plt.xticks(rotation='vertical')
plt.ylabel("F1-micro")
plt.show()
```

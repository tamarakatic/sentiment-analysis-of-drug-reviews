```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
bow=pd.read_csv('../reports/report.csv')
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(bow['model'], bow['f1_micro'])

plt.title("Results of Bag of Words with diff models")
plt.xlabel("Model")
plt.xticks(rotation='vertical')
plt.ylabel("F1-micro")
plt.show()
```

```python
import json
from pprint import pprint

with open("../reports/xgb_report.json", "r") as fp:
    json_report = json.load(fp)

print("Grid Search parameters:")
pprint(json_report['grid_params'])
print("\nBest parameters:")
pprint(json_report['best_params'])
```
